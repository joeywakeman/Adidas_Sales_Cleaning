# Adidas Sales Data Cleaning & Data Wrangling

In this project, I am going to import a messy dataset of Adidas shoe sales information, clean it, and transform it into a dataset suitable for data analysis.

```python
import pandas as pd


pd.set_option("display.max.columns",200)
```

```python
df = pd.read_excel(r"/Users/josephwakeman/Downloads/Adidas US Sales Datasets.xlsx")
df
```

## Renaming the Headers

```python
df1 = df.rename(columns=df.iloc[3])
df1.head()
```

## Dropping Rows & Columns that Imported as Null from Excel

```python
df2 = df1.drop(df1.index[:4])
df2.head()
```

```python
df3 = df2.dropna(axis=1)
df3.head()
```

```python
df4 = df3.reset_index(drop=True)
df4.head()
```

## Updating Data Types

```python
df4.info()
```

```python
df4[["Retailer ID", "Price per Unit", "Units Sold"]] = df4[["Retailer ID", "Price per Unit", "Units Sold"]].astype(int)

df4[["Price per Unit", "Total Sales", "Operating Profit", "Operating Margin"]] = df4[["Price per Unit", "Total Sales", "Operating Profit", "Operating Margin"]].astype(float)

df4.info()
df4.head()
```

## Standardizing Float Format & Adding Thousands Commas

```python
pd.set_option("display.float_format", lambda x: "{:,.2f}".format(x))
df4
```

## Replacing Spaces with Underscores in Headers

```python
df4.columns
```

```python
df5 = df4.rename(columns = {"Retailer ID":"Retailer_ID",
                      "Invoice Date":"Invoice_Date",
                      "Price per Unit":"Price_per_Unit",
                      "Units Sold":"Units_Sold",
                      "Total Sales":"Total_Sales",
                      "Operating Profit":"Operating_Profit",
                      "Operating Margin":"Operating_Margin",
                      "Sales Method":"Sales_Method"})
    
df5.head()
```

## Using a Loop to Remove the Timestamps in the Invoice_Date Column

```python
dates = list()

for x in df5["Invoice_Date"].astype(str):
    y = x.split()
    dates.append(y[0])
    
df5["Invoice_Date"] = dates
df5.head()
```

## Using a Loop to Add Separate Day, Month, and Year Columns, Then Placing Them Next to Invoice_Date

```python
years = list()
months = list()
days = list()

for x in df5["Invoice_Date"]:
    year = x[:4]
    month = x[5:7]
    day = x[8:]
    years.append(year)
    months.append(month)
    days.append(day)
    
df5["Year"] = years
df5["Month"] = months
df5["Day"] = days

df5.head()
```

```python
df5.columns
```

```python
df5 = df5[['Retailer', 'Retailer_ID', 'Invoice_Date', 'Month', 'Day',
       'Year', 'Region', 'State', 'City',
       'Product', 'Price_per_Unit', 'Units_Sold', 'Total_Sales',
       'Operating_Profit', 'Operating_Margin', 'Sales_Method']]
df5.head()
```

## Updating Data Types of New Columns

```python
df5.info()
```

```python
df5[["Year", "Month", "Day"]] = df5[["Year", "Month", "Day"]].astype(int)
df5["Invoice_Date"] = pd.to_datetime(df5["Invoice_Date"])
df5.info()
```
